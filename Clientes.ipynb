Bloque 3 : Limpieza de Datos


```python
#elimina espacio al inicio y al final en columnas de texto
for con in df.select_dtypes(include="object").columns:df[col]=df[col].str.strip()

#normalizar el nombre
df["nombre"]=df["nombre"].str.title()

#normalizar el email
df["email"]=df["email"].str.lower()

#normalizar genero
df["genero"]=df["genero"].str.lower()

#convertir fecha de nacimiento a formato fecha
df["fecha_nacimiento"]=pd.to_datetime(df["fecha_nacimiento"], format='mixed', dayfirst=True)

#Normalizar Ciudad
df["ciudad"]=df["ciudad"].str.title()
df["ciudad"]=df["ciudad"].str.strip()
df["ciudad"]=df["ciudad"].str.upper()
df["ciudad"]=df["ciudad"].str.replace(" ","")

#Limpieza de Segmento
df['segmento']=df['segmento'].astype(str)
df['segmento']=df['segmento'].str.strip()
df['segmento']=df['segmento'].str.upper()
```
